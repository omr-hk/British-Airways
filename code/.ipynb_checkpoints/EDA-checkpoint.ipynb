```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('../Extracted-Data/dataset.xlsx')
data.head()
```

```python
data['REVIEW'].value_counts()
```

It is seen that there are more positive reviews than negative about the airlines

```python
cols=['SEAT_COMFORT', 'CABIN_STAFF_SERVICE',
       'FOOD_AND_BEVERAGES', 'INFLIGHT_ENTERTAINMENT', 'GROUND_SERVICE',
       'WIFI_AND_CONNECTIVITY', 'VALUE_FOR_MONEY', 'RECOMMENDED']
```

```python
plt.hist(data['TYPE_OF_TRAVELLER'].dropna())
plt.title('TYPE OF TRAVELLER')
plt.savefig('../Plots/traveller.png')
```

Majority of the travellers are of the 'Couple Leisure' type, let us now see the majority types under positive and negative reviews.

```python
group=data.groupby('REVIEW')
pos=group.get_group('POSITIVE')
neg=group.get_group('NEGATIVE')
```

```python
plt.hist(pos['TYPE_OF_TRAVELLER'].dropna())
plt.title('TYPE OF TRAVELLER - POSITIVE REVIEWS')
plt.savefig('../Plots/traveller-pos.png')
```

```python
plt.hist(neg['TYPE_OF_TRAVELLER'].dropna())
plt.title('TYPE OF TRAVELLER - NEGATIVE REVIEWS')
plt.savefig('../Plots/traveller-neg.png')
```

It is seen that much of the positive reviews are under 'Solo Leisure' travellers followed by the 'Couple Leisure'. The negative reviews are quite populated under the 'Couple Leisure'travellers followed by the 'Business' travellers. Let us further analyse why that is.

```python
traveler=data.groupby('TYPE_OF_TRAVELLER')
cl=traveler.get_group('Couple Leisure')
sl=traveler.get_group('Solo Leisure')
bs=traveler.get_group('Business')
fl=traveler.get_group('Family Leisure')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(cl[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/couple-leisure.png')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(sl[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/solo-leisure.png')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(bs[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/business.png')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(fl[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/family-leisure.png')
```

```python
route_neg=neg['ROUTE'].value_counts().to_frame().iloc[:20]
route_pos=pos['ROUTE'].value_counts().to_frame().iloc[:20]
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.barh(list(route_neg.index), list(route_neg['ROUTE'].values))
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 5)
ax.grid(color ='grey',
    linestyle ='-.', linewidth = 0.5,
    alpha = 0.2)
ax.invert_yaxis()
ax.set_title('ROUTES - NEGATIVE REVIEWS')
plt.savefig('../Plots/routes-neg.png')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.barh(list(route_pos.index), list(route_pos['ROUTE'].values))
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 5)
ax.grid(color ='grey',
    linestyle ='-.', linewidth = 0.5,
    alpha = 0.2)
ax.invert_yaxis()
ax.set_title('ROUTES - POSITIVE REVIEWS')
plt.savefig('../Plots/routes-pos.png')
```

It is seen that the negative reviews have passengers on the route of London to Johannesburg and not on Johannesburg to London, but the most negative trip reviews seen are on the London to Athens route.

```python
route_group=data.groupby('ROUTE')
lonjoh=route_group.get_group('London to Johannesburg')
athlon=route_group.get_group('Athens to London')
lonath=route_group.get_group('London to Athens')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(lonjoh[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/london-johannesburg.png')
```

It is seen that this route needs to work on its seat comfort,food and beverages,wifi and connectivity and its value for money in order to improve its ratings

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(athlon[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/athens-london.png')
```

```python
fig,ax=plt.subplots(2,4,figsize=(20,10))
x=0
for i in range(0,2):
    for j in range(0,4):
        ax[i][j].hist(lonath[cols[x]])
        ax[i][j].set_title(cols[x])
        x+=1
plt.savefig('../Plots/london-athens.png')
```

It is seen that the Athens to London trip needs an overall haul in their services in order to generate more positive reviews.
